#Plot GDP trends over time for major countries (e.g., line plots for a few selected countries).


```python
countries_to_plot=['China','India','United States','Australia']
filter_data=data[data['Country Name'].isin(countries_to_plot)]

filter_data.set_index('Country Name',inplace=True)
filter_data = filter_data.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code'])
filter_data.head(5)
```

```python
#Transpose the Data
filter_data=filter_data.T
filter_data.head(5)
```

```python
filter_data.plot(figsize=(12,6))
plt.title=("GDP Trends Over time for India,USA,China")
plt.xlabel("Year")
plt.ylabel("GDP (current US$)")
plt.legend(title="Country")
cursor=mplcursors.cursor(hover=True)
@cursor.connect("add")
def on_add(sel):
    year = sel.target[0]  # Get the x-coordinate (year)
    gdp_value = sel.target[1]  # Get the y-coordinate (GDP value)
    sel.annotation.set(text=f"Year: {int(year)}\nGDP: ${gdp_value:,.0f}")
    sel.annotation.get_bbox_patch().set(fc="white", alpha=0.8)  # Set background for better visibility

plt.show()
```
